```python
# Imports necesarios
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#cargamos los datos de entrada

sns.set_style("darkgrid")

path_file = "C:/Users/user/Desktop/pipe/Proyecto/Datos_Generales/Apendice_E.xlsx"

dta = pd.read_excel(path_file)
df = pd.DataFrame(dta)
```

```python
#veamos cuantas dimensiones y registros contiene
dta.shape
```

```python
# Ahora veamos algunas estadísticas de nuestros datos
dta.drop(['PUBLICI'], axis=1, inplace=True)
dta
```

```python
X = pd.DataFrame(dta.iloc[:,0:8].values)
print(X)

y = dta.iloc[:,[8]]
print(y)
```

```python
reg = LinearRegression()
regR = reg.fit(X, y)

y_pred = regR.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
r2 = reg.score(X, y)
```

```python
print("El error es: ", error)
print("El valor de r^2 es: ", r2)
print("Los coeficientes son: ", reg.coef_)
```
