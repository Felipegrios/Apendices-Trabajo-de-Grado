```python
X = pd.DataFrame(dta.iloc[:,0:9].values)
print(X)

y = dta.iloc[:,[9]]
print(y)
```

```python
gamma_model = sm.GLM(y,X)
gamma_resul = gamma_model.fit()
print(gamma_resul.summary())

y_pred = gamma_resul.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y,y_pred)

print("El error es: ", error)
print("El valor de r^2 es: ", r2)
```

```python
poisson_model = sm.Poisson(y, X)  
poisson_res = poisson_model.fit(method="newton")
print(poisson_res.summary())

y_pred = poisson_res.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("El error es: ", error)
print("El valor de r^2 es: ", r2)
```
