```python
X = pd.DataFrame(dta.iloc[:,0:8].values)
print(X)
y = dta.iloc[:,[8]]
print(y)
```

```python
XX = pd.DataFrame(dta_2018.iloc[:,0:8].values)
print(XX)
yy = dta_2018.iloc[:,[8]]
print(yy)
```

```python
reg = LinearRegression()
reg = reg.fit(X, y)

y_pred = reg.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
r2 = reg.score(X, y)

print("El error es: ", error)
print("El valor de r^2 es: ", r2)
print("Los coeficientes son: ", reg.coef_)
```

```python
y_pred2 = reg.predict(XX)
error2 = np.sqrt(mean_squared_error(yy, y_pred2))
r22 = reg.score(XX, yy)
print("El error es: ", error2)
print("El valor de r^2 es: ", r22)
```
